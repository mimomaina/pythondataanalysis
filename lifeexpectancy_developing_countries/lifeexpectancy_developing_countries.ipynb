```python
# Filter for developing countries
developing = life01[life01['Status'] == 'Developing']

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = developing['Life_expectancy'].quantile(0.25)
Q3 = developing['Life_expectancy'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outlier countries based on life expectancy
outliers = developing[(developing['Life_expectancy'] < lower_bound) | (developing['Life_expectancy'] > upper_bound)]

# Extract the country names of outliers
outlier_countries = outliers['Country'].tolist()

# Load world map shapefile
world = gpd.read_file(gpd.datasets.get_path('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_110m_admin_0_sovereignty.geojson'))

# Filter for developing countries in the world shapefile (assuming 'status' column exists)
world_developing = world[world['name'].isin(outlier_countries)]

# Plot the world map
fig, ax = plt.subplots(figsize=(10, 8))
world.plot(ax=ax, color='lightgray', edgecolor='black')  # plot the entire world map

# Highlight the outlier countries
world_developing.plot(ax=ax, color='red', edgecolor='black')  # highlight outliers in red

# Add labels for the outlier countries
for idx, row in world_developing.iterrows():
    ax.annotate(row['name'], xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                xytext=(3, 3), textcoords="offset points", fontsize=10, color='black')

# Title and show the plot
plt.title("Outlier Countries Based on Life Expectancy", fontsize=15)
plt.show()
```
